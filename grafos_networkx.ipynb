# GRAFO SIMPLES
---

### Somatório dos Graus dos vértice
```
Σ d(v) = 2 * |E|
```

* Σ = somatório
* d(v) = grau de um grafos (somas dos graus de seus vértices) qualquer
> grau de um vértice = número de aretas que incidem nele.
* E = arestas



````
# Instancia um grafo simples.
G = nx.Graph()

# Limpa o grafo.
G.clear()

# 4 nós (vértices) numa lista.
G.add_nodes_from([1, 2, 3, 4])

# 3 arestas numa lista.
G.add_edges_from([(1, 2), (1, 3), (1, 4)])

# Número de nós (vértices).
G.number_of_nodes()

# Número de arestas.
G.number_of_edges()

# Desenhe o gráfico.
nx.draw(G, with_labels=True, font_weight='bold', node_size=1000)

# Coloque informações sobre as conecções realizadas pelos grafos.
G[1][2]['ligação'] = "com grafo 2"
G[1][3]['ligação'] = "com grafo 3"

# Exiba essas informações.
G[1]
````

# GRAFO REGULAR (r-regular)
---
### Somatório dos graus de um grafo regular
```
Σ d(v) = n*r
```

* Σ = somatório
* d(v) = grau de um grafos (somas dos graus de seus vértices) qualquer
* n = vértices
* r = número do grau regular (r-regular)

---

### Cálculo do número de arestas num grafo regular (m)
```
m = (n*r)/2
```
> somatório dos graus de um grafo -> Σ d(v) = 2m  
> somatório dos graus de um grafo regular -> Σ d(v) = n*r  
> Logo, 2m = n*r.  
> Assim, m = (n * r)/2

* m = arestas
* n = vértices
* r = r-regular

````
# Variável recebe o valor regular de grau r (r-regular).
r_regular = 4

# Variável recebe o valor do número de vértices que constituem o grafo.
vertice = 6

# Geração randômica de grafo regular.
grafo = nx.generators.random_regular_graph(r_regular, vertice)

# Desenho (plot) do grafo em tela.
nx.draw(grafo, with_labels=True)
````

# GRAFO COMPLETO (Kn)
---

### Arestas num grafo completo (Kn)
```
|E(Kn)| = Cn,2 = (n^2 - n)/2
```
* E = arestas
* Kn = grafo completo
* C = fatorial
* n = vértice

````
# Gera um grafo completo.
grafo_completo_kn = nx.complete_graph(6)

# Desenha (plota) o grafo.
nx.draw(grafo_completo_kn)
````

# GRAFO BIPARTIDO OU BIPARTITE
---

````
# Instancia um grafo simples.
B = nx.Graph()

# Adição de nó bipartido.
B.add_node(1, bipartite=0)

# Adição de nós bipartidos.
B.add_nodes_from([n for n in range(1, 30)], bipartite=1)

# Adição de arestas, note que o nó 1 realiza cominucação com os outros nós.
B.add_edges_from([(1, 2), (1, 3), (1, 4), (1, 5), (1, 6), (1, 7), (1, 8), 
                  (1, 9), (1, 10), (1, 11), (1, 12), (1, 13), (1, 14), (1, 15), 
                  (1, 16), (1, 17), (1, 18), (1, 19), (1, 20), (1, 21), (1, 22),
                  (1, 23), (1, 24), (1, 25), (1, 26), (1, 27), (1, 28), (1, 29)])

# Desenha (plota) o grafo.
nx.draw(B, with_labels=True)

# Retorna True se o grafo for bipartite.
bipartite.is_bipartite(B)
````

# GRAFO BIPARTIDO COMPLETO
---
### Total de vértices
```
n = p + q
```
* n = vértices
* p = subconjunto p
* q = subconjunto q
---
### Total de arestas
```
m = p * q
```
* m = arestas
* p = subconjunto p
* q = subconjunto q
---


````
# Variáveis que armazenam os subconjuntos que compõem o grafo bipartido.
subconjunto_p = 3  
subconjunto_q = 5

# Gera um grafo completo.
grafo_bipartido_completo_kpq = nx.complete_bipartite_graph(subconjunto_p, subconjunto_q)

# Desenha (plota) o grafo.
nx.draw(grafo_bipartido_completo_kpq, with_labels=True)

# Correspondência máxima da cardinalidade de um grafo bipartito completo.
bipartite.maximum_matching(grafo_bipartido_completo_kpq)

# nx.draw(bipartite.random_graph(3, 5, 0.6), with_labels=True)
````
